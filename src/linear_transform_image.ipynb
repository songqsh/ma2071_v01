```
# load this image as numpy array, and see it.
img = cv2.imread('wpilogo.jpg')
print(img.shape)
plt.imshow(img)
```

```
# the image size is recorded in rows, cols
rows,cols, _ = img.shape
print(rows, cols)
```

- Linear Transform 1

```
lintrans = [[.5, 0], [0, 0.5]]
shift = [100, 100]
M = np.append(np.transpose(lintrans), [shift], axis = 0).transpose()

img1 = cv2.warpAffine(img, M, (cols,rows)) #package run your instruction
plt.imshow(img1)
```

- Linear Transform 2

```
lintrans = [[.5, 1.], 
            [0., 1.]] #this is linear transformation, where you need to replace
shift = [-80., 0.]
M = np.append(np.transpose(lintrans), [shift], axis = 0).transpose()

img2 = cv2.warpAffine(img1, M, (cols,rows)) #package run your instruction
plt.imshow(img2)
```
