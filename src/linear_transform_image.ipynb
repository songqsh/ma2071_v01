```
# load this image as numpy array, and see it.
img = cv2.imread('wpilogo.jpg')
print(img.shape)
plt.imshow(img)
```

```
# the image size is recorded in rows, cols
rows,cols, _ = img.shape
print(rows, cols)
```

Next are some linear transformations. Can you explain?

```
lintrans = [[.8, 1.6], 
            [0., 1.6]] #this is linear transformation, where you need to replace

shift = [0., 0.]
M = np.append(np.transpose(lintrans), [shift], axis = 0).transpose()

res = cv2.warpAffine(img, M, (2*cols,2*rows)) #package run your instruction
plt.imshow(res)
```
