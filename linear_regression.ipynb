Linear regression

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data points
x = np.array([2, 5, 7, 8])
y = np.array([1, 2, 3, 3])
print(f'x = {x}')
print(f'y = {y}')
```

```python
# plot the points in x-y plane
plt.scatter(x, y)
plt.show()
```

```python
# assign matrix
X = np.array([np.ones(len(x)), x]).T
print(f'X = {X}')
```

```python
#calculate the solution
beta = np.linalg.inv(X.T @ X) @ X.T @ y
print(f'beta = {beta}')
```

```python
# plot the line through the scattered dots
plt.scatter(x, y)
plt.plot(x, beta[0] + beta[1] * x, 'r')
plt.show()
```
